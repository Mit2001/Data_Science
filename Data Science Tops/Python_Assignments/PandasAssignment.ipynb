```python
# Read the .tsv file
chipo = pd.read_csv("C:\\Users\\mitps\\Downloads\\chipotle.tsv", delimiter='\t')

# Save as .csv file
chipo.to_csv("C:\\Users\\mitps\\Downloads\\chipo.csv", index=False)
```

```python

chipo.head()
```

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', ' ').astype(float)
cost = chipo.groupby('item_name')['item_price'].sum()
items_above_10 = cost[cost > 10.00]
count_items_above_10 = len(items_above_10)
print(count_items_above_10)
```

```python
# What is the price of each item?
# print a data frame with only two columns item_name and item_price


cost = chipo.groupby('item_name')['item_price'].mean()
cost.columns = ['item_name', 'item_price']
print(cost)
```

```python
#Sort by the name of the item
cost = chipo.groupby('item_name')['item_price'].mean().reset_index()
sorted_items = cost.sort_values('item_name')
print(sorted_items)
```

```python
#What was the quantity of the most expensive item ordered?
item = chipo.loc[chipo['item_price'].idxmax()]
quantity_expensive = item['quantity'] 
print(item,quantity_expensive)
# # Get the quantity of the most expensive item ordered
# quantity_most_expensive = most_expensive_item['quantity']
```

```python
# How many times was a Veggie Salad Bowl ordered?
veggie_salad_bowl_orders = (chipo['item_name'] == 'Veggie Salad Bowl').sum()
print(veggie_salad_bowl_orders)
```

```python
# How many times did someone order more than one Canned Soda?
canned_soda_multiple_orders = ((chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)).sum()
print(canned_soda_multiple_orders)
```

```python
import pandas as pd

# Example: Read a CSV file from a URL
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'
euro12 = pd.read_csv(url)

# Print the fetched data
euro12.head()
```

```python
# Select only the Goal column 
Goal_Column = euro12['Goals']
Goal_Column
```

```python
# How many team participated in the Euro2012? 
Team_Participate = euro12['Team'].nunique()
Team_Participate
```

```python
# What is the number of columns in the dataset? 
euro12.shape
euro12.shape[1]
```

```python
# View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print(discipline)
```

```python
# Sort the teams by Red Cards, then to Yellow Cards 
x = euro12[['Red Cards','Yellow Cards']]
x
sort1 = x.sort_values(['Red Cards', 'Yellow Cards'])
sort1
```

```python
# Calculate the mean Yellow Cards given per Team 
ycmean = euro12['Yellow Cards'].mean() 
print(ycmean)
```

```python
# Filter teams that scored more than 6 goals 
morethansixgoals = euro12[euro12['Goals']>6]
print(morethansixgoals)
```
