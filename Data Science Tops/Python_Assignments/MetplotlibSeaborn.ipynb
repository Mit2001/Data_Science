# Liver_disease_data

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("C:\\Users\\mitps\\Downloads\\Liver_disease_data.csv")
df.head()
```

```python
null_value = df.isnull().sum()
null_value
```

```python
df.tail()
```

```python
avg_total_alcohol_consumption = df.groupby('Gender')['AlcoholConsumption'].mean()
plt.figure(figsize=(10,6))
avg_total_alcohol_consumption.plot.bar() 
plt.title("avg_total_alcohol_consumption")
plt.xlabel('Gender')

plt.ylabel('AlcoholConsumption')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['LiverFunctionTest'], bins = 20, edgecolor = 'black')
# plt.hist(titanic_data['Age'], bins=20, color='blue', edgecolor='black')
plt.xlabel('LiverFunctionTest') 
# plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Liver Function Test Values')
# plt.grid(True)
plt.show()
```

```python
# Pie chart of Diagnosis distribution
# diagnosis_counts = data['Diagnosis'].value_counts()
diagnosis_counts = df['Diagnosis'].value_counts() 
plt.figure(figsize=(8, 8))
explode = [0.2,0]
plt.pie(diagnosis_counts, labels = diagnosis_counts.index, autopct='%1.1f%%', startangle=140, explode = explode)
# plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title('Pie Chart of Diagnosis Distribution')
plt.axis('equal')
plt.show()
```

```python
df.head()
liver = df.groupby('Diagnosis')['LiverFunctionTest'].mean() 
plt.figure(figsize=(10,6))
liver.plot.barh() 
plt.title('Avg Liver Function Test By Diagnosis')
plt.xlabel('LiverFunctionTest')
plt.show()
```

```python


# Calculate gender counts
gender_counts = df['Gender'].value_counts()

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Bar plot
ax[0].bar(gender_counts.index, gender_counts.values, color='k')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Number of Records')
ax[0].set_title('Number of Records by Gender')

# Horizontal bar plot
ax[1].barh(gender_counts.index, gender_counts.values, color='k')
ax[1].set_ylabel('Gender')
ax[1].set_xlabel('Number of Records')
ax[1].set_title('Number of Records by Gender')

# Show plots
plt.tight_layout()
plt.show()
```
